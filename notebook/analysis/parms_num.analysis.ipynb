```python
import os
import seaborn as sns
import pandas as pd
import numpy as np

sns.set_theme(style="darkgrid")
path_dt = {
    'gm12878_1kb': r"gm12878_ctcf-1000bp/result_all.csv",
    'gm12878_3kb': r"gm12878_ctcf-3000bp/result_all.csv",
    'helas3_1kb': r"helas3_ctcf-1000bp/result_all.csv",
    'helas3_3kb': r"helas3_ctcf-3000bp/result_all.csv"
}
```

```python
# 中文路徑問題，用絕對路徑
root = "C:/Users/aaron/Project/碩論/code/benchmark/output/experiment/2024.08.26__MCIENet_final/" 
```

```python
cols = ['Hl', 'Hs',	'CHs', 'phase', 'auPRCs', 'total', 'extractor', 'classifier', 'ext_ratio',
       'clf_ratio']
df = pd.read_csv(os.path.join(root, path_dt['gm12878_1kb']), usecols=cols)
df['total(log)'] = np.log(df['total'])
df
```

```python
df_test = df[df['phase'] == 'test']
df_test['Vanishing Grad'] = df_test['auPRCs'] < 0.3
```

```python
g = sns.jointplot(x="total", y="auPRCs", data=df_test,
                  kind="reg", color="m", height=7)
```

```python
sns.relplot(
    data=df_test, kind="line",
    x="total", y="auPRCs", hue='Vanishing Grad',
    col="Hl", 
    estimator=None
)
```
