```python
import os
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")

def draw_heatmap(df, method='spearman'):
    # Compute the correlation matrix
    corr = df.corr(method=method) # ‘pearson’, ‘kendall’, ‘spearman’

    # Generate a mask for the upper triangle
    mask = np.triu(np.ones_like(corr, dtype=bool))

    # Set up the matplotlib figure
    f, ax = plt.subplots(figsize=(11, 9))

    # Generate a custom diverging colormap
    cmap = sns.diverging_palette(230, 20, as_cmap=True)

    # Draw the heatmap with the mask and correct aspect ratio
    plot = sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, vmin=-1, center=0,
                square=True, linewidths=.5)

    return plot

path_dt = {
    'gm12878_1kb': r"gm12878_ctcf-1000bp/result_all.csv",
    'gm12878_3kb': r"gm12878_ctcf-3000bp/result_all.csv",
    'helas3_1kb': r"helas3_ctcf-1000bp/result_all.csv",
    'helas3_3kb': r"helas3_ctcf-3000bp/result_all.csv"
}
# 中文路徑問題，用絕對路徑
root = r"I:\我的雲端硬碟\Bio_Lab\碩論\code\benchmark\output\experiment\2024.08.26__MCIENet_final/" 
```

```python
cols = ['Hl', 'Hs',	'CHs', 'phase', 'auPRCs', 'total', 'extractor', 'classifier', 'ext_ratio',
       'clf_ratio']
df = pd.read_csv(os.path.join(root, path_dt['gm12878_1kb']), usecols=cols)
df_test = df[df['phase'] == 'test']
df_test.drop('phase', axis=1, inplace=True)
df_test.head()

df_test_parms = df_test[['total', 'extractor', 'classifier', 'auPRCs']]
df_test_eval = df_test[['Hl', 'Hs', 'CHs', 'auPRCs']]
```

```python
draw_heatmap(df_test_parms, method='spearman')
```

```python
draw_heatmap(df_test_eval, method='spearman')
```
